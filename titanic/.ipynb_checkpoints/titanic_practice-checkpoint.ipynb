```python
#数据选择类
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import FeatureUnion
```

```python
all_data = train_data.append(test_data,ignore_index=True)

#名字分解出头衔
title_map={
        'Capt':'Officer',
        'Col':'Officer',
        'Major':'Officer',
        'Jonkheer':'Royalty',
        'Don':'Royalty',
        'Sir':'Royalty',
        'Dr':'Officer',
        'Rev':'Officer',
        'the Countess':'Royalty',
        'Mme':'Mrs',
        'Mlle':'Mr',
        'Ms':'Mrs',
        'Mr':'Mr',
        'Mrs':'Mrs',
        'Miss':'Miss',
        'Master':'Master',
        'Lady':'Royalty'
    }


def gettitles(name):
    return name.split(',')[1].split('.')[0].strip()

all_data['Name']=all_data['Name'].map(gettitles).map(title_map)

#Cabin首字母提取
all_data['Cabin']=all_data['Cabin'].str[:1]
```

```python
all_data.head
```

```python
#ticket数据删除
all_data.drop(['Ticket'],axis=1,inplace=True)
```

```python
all_data.info()
```

```python
#age填充分组均值
all_mean =all_data.groupby(['Sex','Pclass','Name']).mean().sort_values(by=['Sex','Pclass','Name']).reset_index()[['Sex','Pclass','Name','Age']]    

def fill_age(row):
    condition = (
    (row['Sex']==all_mean['Sex'])&
    (row['Pclass']==all_mean['Pclass'])&
    (row['Name']==all_mean['Name'])
    )
    return all_mean[condition]['Age'].values[0]

all_data['Age'] =all_data.apply(lambda row: fill_age(row) if np.isnan(row['Age']) else row['Age'], axis=1)
```

```python
all_data
```

```python
#构建流水线
num_pipeline = Pipeline([
        ('select_num', DataFrameSelector(['Age','SibSp','Parch'])),
        ('imputer', SimpleImputer(strategy='constant',fill_value=0)),
    ])
#Cabin缺失值填充/OneHot编码处理
cabin_pipeline = Pipeline([
        ('select_obj', DataFrameSelector(['Cabin'])),
        ('imputer', SimpleImputer(strategy='constant',fill_value='U')),
        ('cat_encoder', OneHotEncoder(sparse=False)),
    ])
#Fare缺失值填充
fare_pipeline = Pipeline([
        ('select_num', DataFrameSelector(['Fare'])),
        ('imputer', SimpleImputer(strategy='constant',fill_value=14.6)),
    ])
#OneHot编码处理
cat_pipeline = Pipeline([
        ('select_cat', DataFrameSelector(['Pclass', 'Sex', 'Embarked','Name'])),
        ('imputer', SimpleImputer(strategy='constant',fill_value='C')),
        ('cat_encoder', OneHotEncoder(sparse=False)),
    ])
from sklearn.pipeline import FeatureUnion
preprocess_pipeline = FeatureUnion(transformer_list=[
        ('num_pipeline', num_pipeline),
        ('cabin_pipeline', cabin_pipeline),
        ('fare_pipeline', fare_pipeline),
        ('cat_pipeline', cat_pipeline),
    ])
```

```python
X_all=preprocess_pipeline.fit_transform(all_data)
X_train = X_all[:891,:]
X_test=X_all[891:,:]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
#网格搜索

param_grid={
    'max_features':['auto'],
    'n_estimators':[45,46,47],
    'min_samples_leaf':[8,9,10],
    #'min_samples_split':[5,8,10,12],
    'max_depth':[8,9,10],
    'oob_score':['True']
}
forest_reg = RandomForestClassifier()
cross_validation = StratifiedKFold(n_splits=5)

grid_search=GridSearchCV(
    forest_reg,
    param_grid ,
    scoring='accuracy',
    cv = cross_validation,
    verbose=1
)
# 训练模型
grid_search.fit(X_train,y_train)
model = grid_search
parameters = grid_search.best_params_
'''
forest_reg = RandomForestClassifier()
forest_scores = cross_val_score(forest_reg, X_train, y_train, cv=10)
forest_scores.mean()
'''
```
